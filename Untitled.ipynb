```python
from torch_geometric.nn import fps, knn_graph
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
n=3000

alpha = 1
a = np.random.uniform(size=2*n, low=-3, high = 3).reshape((-1, 2))
a = a[a[:,0]<alpha/a[:,1]]
a = a[a[:,0]> -alpha/a[:,1]]
a = torch.tensor(a)
```

```python
plt.scatter(a[:, 0], a[:, 1])
```

```python
a = a[fps(a, ratio=0.4)]
```

```python
plt.scatter(a[:, 0], a[:, 1])
```
